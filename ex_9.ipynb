```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Name': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E'],
    'Stock': [120, 80, 150, 60, 200]
})
df = pd.read_csv("products.csv")
```

```python
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt


plt.plot(df['Name'], df['Stock'], marker='o', linestyle='-', color='blue')
plt.xlabel('Product Name')
plt.ylabel('Stock Quantity')
plt.title('Line Chart of Product Stock')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


plt.bar(df['Name'], df['Stock'], color='green')
plt.xlabel('Product Name')
plt.ylabel('Stock Quantity')
plt.title('Bar Chart of Product Stock')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


plt.scatter(df['Name'], df['Stock'], color='red')
plt.xlabel('Product Name')
plt.ylabel('Stock Quantity')
plt.title('Scatter Plot of Product Stock')
plt.show()
```
