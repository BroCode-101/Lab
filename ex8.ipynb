```python
import matplotlib.pyplot as plt

x = [0,1,2,3,4,5]
y = [0,1,4,9,16,25]
plt.plot(x,y,marker='o',linestyle='-',color='blue')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()

```

```python
x  = [1,2,3,4,5]
y1 = [1,2,3,4,5]
y2 = [1,4,9,16,25]
y3 = [5,4,3,2,1]

plt.plot(x,y1,label="linear",color="green",linewidth=2)
plt.plot(x,y2,label="quadratic",color ="blue",linewidth=3)
plt.plot(x,y3,label="reverse",color = "red",linewidth=1)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Multiple Lines with Legends')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]
y3 = [5, 4, 3, 2, 1]

plt.plot(x, y1, label='Solid', linestyle='-', color='purple')
plt.plot(x, y2, label='Dashed', linestyle='--', color='orange')
plt.plot(x, y3, label='Dotted', linestyle=':', color='cyan')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Multiple Line Styles')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt


x = []
y = []

with open('test.txt', 'r') as file:
    for line in file:
        values = line.strip().split()
        x.append(float(values[0]))
        y.append(float(values[1]))


plt.plot(x, y, marker='o', color='magenta')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot from File Data')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt


languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]


plt.bar(languages, popularity, color='skyblue')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Popularity of Programming Languages')
plt.show()
```

```python
import matplotlib.pyplot as plt

languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.barh(languages, popularity, color='lightgreen')
plt.xlabel('Popularity (%)')
plt.ylabel('Programming Languages')
plt.title('Popularity of Programming Languages (Horizontal)')
plt.show()
```

```python
import matplotlib.pyplot as plt


languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.figure(figsize=(8, 6))
plt.pie(popularity, labels=languages, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Popularity of Programming Languages - Pie Chart')
plt.axis('equal')  # Ensures pie is drawn as a circle
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50) * 100  # Random values between 0 and 100
y = np.random.rand(50) * 100

plt.scatter(x, y, color='teal', marker='o')
plt.xlabel('Random X')
plt.ylabel('Random Y')
plt.title('Scatter Plot of Random Distribution')
plt.grid(True)
plt.show()
```
